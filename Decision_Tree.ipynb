```python
import pandas as pd
```

```python
import requests
import io

url = "https://raw.githubusercontent.com/codebasics/py/master/ML/9_decision_tree/salaries.csv"

download = requests.get(url).content

df = pd.read_csv(io.StringIO(download.decode('utf-8')))
```

```python
df
```

```python
inputs = df.drop(['salary_more_then_100k'],axis = 'columns')
```

```python
targets = df['salary_more_then_100k']
```

```python
inputs
```

```python
targets
```

```python
# As we know machine learning algorithms only works with numbers so we will do labeling
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# so we will do encoding of 3 columns here company , job and degree
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company_n']=le_company.fit_transform(inputs['company'])
inputs['job_n']=le_job.fit_transform(inputs['job'])
inputs['degree_n']=le_degree.fit_transform(inputs['degree'])
```

```python
inputs
```

```python
# remove the extra columns now
```

```python
inputs_n = inputs.drop(['company','job','degree'],axis='columns')
```

```python
inputs_n
```

```python
# now we will split the dataset
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test =train_test_split(inputs_n,targets,test_size=0.2)
```

```python
X_train
```

```python
X_test
```

```python
# now we are using DecisionTreeClassifier
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict([[2,2,1]]) #the person who is working in google as a sales executive and have done masters will paid less than 100k 
```
